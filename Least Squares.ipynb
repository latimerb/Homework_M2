# Method of Least Squares (Linear Regression)
In this notebook, you will try to fit a line to data like a computer would do it, by optimizing a cost function. Let's begin.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
```

First let's create a line and then add some random noise to construct a dataset.

```python
x = np.array([0,1,2,3,4])
y = 2*x
```

```python
print y
```

```python
plt.plot(x,y)
```

```python
y_real = y + np.random.randn(1,5)
plt.plot(x,y)
plt.plot(x,y_real[0,:],'r.')
```

## Fit a line and plot the cost function
Let's take some "guesses" at what the best fit line will be (we already know it is y = 2x). In the following formulation, the slope of the line is theta_1 and the y-intercept is theta_0. We will assume theta_0 to be zero and makes guesses about theta_1.

```python
theta_0 = 0
theta_1 = 6

y_hat = theta_1*x + theta_0

# Plot the "real" data as red dots
plt.plot(x,y_real[0,:],'r.')
# Plot the "guess" as a line
plt.plot(x,y_hat)
```

You can see clearly that this is a pretty bad guess. Let's calculate the cost. You know the cost function:

$$J = \frac{\sum_{n=1}^{M} (\hat{y} - y)^2)}{2M}$$

```python
M = len(x)
cost = (1/(2.0*M))*np.sum((y_real-y_hat)**2)
print(cost)
```

Now that we've calculated the cost for our guess, let's store it in a vector called J.

```python
J = np.zeros((10,1))
J[6] = cost
```

Let's visualize the cost. All of the costs except the sixth will be zero for now since we haven't calculated them.

```python
plt.plot(np.arange(0,10),J,'bx')
```

We can run the above algorithm interatively using a for loop:

```python
for i in np.arange(0,10):

    theta_0 = 0
    # theta_1 will be 0,1,2,3,4,5...9
    theta_1 = i

    y_hat = theta_1*x + theta_0

    cost = (1/(2.0*M))*np.sum((y_real-y_hat)**2)
    
    J[i] = cost
```

```python
plt.plot(x,y_real[0,:],'r.')
# Plot a few of our guesses
plt.plot(x,J[0]*x)
plt.plot(x,J[2]*x)
plt.plot(x,J[3]*x)
```

```python
plt.plot(np.arange(0,10),J,'bx')
```

You can clearly see that the minimum of the cost function occurs at theta_1 == 2, which makes sense because that is the line we used to construct the data!
